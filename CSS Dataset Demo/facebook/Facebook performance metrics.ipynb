```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("dataset_Facebook.csv")

data
```

```python
# Summary statistics for numeric columns
numeric_cols = ['Page total likes', 'Post Month', 'Post Weekday', 'Post Hour',
                'Lifetime Post Total Reach', 'Lifetime Post Total Impressions', 'Lifetime Engaged Users']
print(data[numeric_cols].describe())

```

```python
# Distribution of post types
post_type_counts = data['Type'].value_counts()
plt.bar(post_type_counts.index, post_type_counts.values)
plt.xlabel('Post Type')
plt.ylabel('Count')
plt.title('Distribution of Post Types')
plt.show()

```

```python
# Average reach for each post type
avg_reach_by_type = data.groupby('Type')['Lifetime Post Total Reach'].mean()
avg_reach_by_type.plot(kind='bar')
plt.ylabel('Average Reach')
plt.title('Average Reach by Post Type')
plt.show()
```

```python
# Heatmap of correlations between numeric variables
correlation_matrix = data[numeric_cols].corr()
plt.figure(figsize=(10, 8))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(numeric_cols)), numeric_cols, rotation=45)
plt.yticks(range(len(numeric_cols)), numeric_cols)
plt.title('Correlation Heatmap')
plt.show()
```
